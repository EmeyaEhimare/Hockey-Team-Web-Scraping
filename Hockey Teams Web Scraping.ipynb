```python
table = soup.find(class_ = 'table')
```

```python
hockey_team_df  = {
    'Team Name' : [],
 'Year': [],
 'Wins': [],
 'Losses' : [],
 'OT Losses' : [],
 'Win %' : [],
 'Goals For (GF)': [],
 'Goals Against (GA)' : [],
 '+ / -': []
}


for i in range(1,26):
    url = f'https://www.scrapethissite.com/pages/forms/?page_num={i}&per_page=25'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    table = soup.find(class_ = 'table')
    table_data = table.find_all('tr', class_ = 'team')




    for data in table_data:
            team_name = data.find('td',class_ = 'name').text.strip()
            year = data.find('td', class_ = 'year').text.strip()
            wins =  data.find('td', class_ = 'wins').text.strip()
            losses = data.find('td', class_ = 'losses').text.strip()
            ot_losses =  data.find('td', class_ = 'ot-losses').text.strip()
            win_percent = data.find('td', class_ = ['pct text-success', 'pct text-danger']).text.strip()
            goals_for = data.find('td', class_ = 'gf').text.strip()
            goals_against = data.find('td', class_ = 'ga').text.strip()
            plus_or_minus = data.find('td', class_ = "diff text-success")
            
            
            if plus_or_minus is not None:
                plus_or_minus = plus_or_minus.text.strip()
            else:
                plus_or_minus = "N/A"  
        
            hockey_team_df['Team Name'].append(team_name)
            hockey_team_df['Year'].append(year)
            hockey_team_df['Wins'].append(wins)
            hockey_team_df['Losses'].append(losses)
            hockey_team_df['OT Losses'].append(ot_losses)
            hockey_team_df['Win %'].append(win_percent)
            hockey_team_df['Goals For (GF)'].append(goals_for)
            hockey_team_df['Goals Against (GA)'].append(goals_against)
            hockey_team_df['+ / -'].append(plus_or_minus)

```

```python
import pandas as pd 
```

```python
df = pd.DataFrame(hockey_team_df)
```
